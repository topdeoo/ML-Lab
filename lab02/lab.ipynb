## 数据预处理

我们知道，感知机只能作用于线性可分的数据集上，于是我们需要确定我们这个数据集是否线性可分

```python
df = pd.read_csv('Data.csv')
df.head(5)
df = df.drop(['id'], axis=1)
```

这里我们画出不同特征间的散点图，期望找到两种特征使得这个数据集是线性可分的（肉眼寻找，而且为了减小难度，直接把特征减到两维了）

```python
for i in df.columns:
    for j in df.columns:
        if i != j and i != 'class' and j != 'class':
            sns.relplot(x=i, y=j, hue='class', data=df)
```

这里选择了前两个特征，也就是 `sepal length (cm)`，`sepal width (cm)` 这两个特征，然后由于感知机只能坐二分类，所以我们我们将Setosa鸢尾花视为正类，其他两种视为负类。

```python
X = df.drop(['class', 'petal length (cm)', 'petal width (cm)'], axis=1).to_numpy()
Y = df['class'].to_numpy()

for i in range(Y.shape[0]):
    if Y[i] == 0:
        Y[i] = 1
    else:
        Y[i] = -1
train_X, test_X, train_Y, test_Y = train_test_split(X, Y, test_size=0.2, shuffle=True)
```
